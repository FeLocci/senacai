```python
#primeira versão
import numpy as np
import pandas as pd

# --- 1. modelando o problema ---

# Func de Ativação Sigmóide
# Transformado resultado da soma ponderada em um valor entre 0 e 1.
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivada da Sigmóide
# usando no Backpropagation.
def sigmoid_derivative(x):
    return x * (1 - x)

# --- 2. preparando os dados com a tabela ---

# XOR em um DataFrame do Pandas
dados_xor = pd.DataFrame({
    'Entrada_1': [0, 0, 1, 1],
    'Entrada_2': [0, 1, 0, 1],
    'Resultado_Esperado': [0, 1, 1, 0]
})

print("Tabela Verdade do Desafio XOR:")
print(dados_xor)
print("-" * 30)

# Convert matrizes NumPy (aplicando cálculos matemáticos)
X = dados_xor[['Entrada_1', 'Entrada_2']].values
y = dados_xor[['Resultado_Esperado']].values

# --- 3. CONSTRUÇÃO  (Backpropagation) ---

def treinar_rede_neural(entradas, alvos, epocas=15000, lr=0.5):
    # Inicializando Hiperparâmetros
    input_nodes = 2
    hidden_nodes = 3  # Camada oculta resolve a não-linearidade do XOR
    output_nodes = 1

    # Inicialização aleatória de pesos
    # pesos_0: liga a entrada a camada oculta
    # pesos_1: liga a camada oculta a saída
    pesos_0 = np.random.uniform(size=(input_nodes, hidden_nodes))
    pesos_1 = np.random.uniform(size=(hidden_nodes, output_nodes))

    for i in range(epocas):
        # FEEDFORWARD
        camada_escondida = sigmoid(np.dot(entradas, pesos_0))
        camada_saida = sigmoid(np.dot(camada_escondida, pesos_1))

        # BACKPROPAGATION

        # 1. erro na saída
        erro_saida = alvos - camada_saida

        # 2. Delta de Saída  (ajuste)
        delta_saida = erro_saida * sigmoid_derivative(camada_saida)

        # 3. Calculando erro da camada oculta
        erro_escondida = delta_saida.dot(pesos_1.T)
        delta_escondida = erro_escondida * sigmoid_derivative(camada_escondida)

        # 4. atualizando pesos (Gradiente Descendente)
        pesos_1 += camada_escondida.T.dot(delta_saida) * lr
        pesos_0 += entradas.T.dot(delta_escondida) * lr

    return pesos_0, pesos_1, camada_saida

# --- 4.resultado

# Chamada da função de treino
p0, p1, predicoes_finais = treinar_rede_neural(X, y)

# Criando tabelas para comparar os resultado
resultados = dados_xor.copy()
resultados['Predicao_IA'] = predicoes_finais
resultados['Erro_Final'] = abs(resultados['Resultado_Esperado'] - resultados['Predicao_IA'])

print("\nResultados Finais após Treinamento:")
print(resultados.to_string(index=False))

# Validando sucesso
print("\nConclusão:")
for p in predicoes_finais:
    status = "Correto" if (p > 0.5 and 1) or (p <= 0.5 and 0) else "Incorreto"
    print(f"Valor predito: {p[0]:.4f} -> Interpretação: {round(p[0])}")
```

```python
#2ª versao
import numpy as np
#Função de ativação (sigmoide) e sua derivada
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
  return x * (1 - x)

#preparando os dados
# Entrada: A, B

inputs = np.array([[0,0], [0,1], [1,0], [1,1]])
#saida esoerada

expected_output = np.array([[0], [1], [1], [0]])

#Iniciando pesos e BIAS (aleatorios)
inputLayerNeurons, hiddenLayerNeurons, outputLayerNeurons = 2, 2, 1

#Pesos entrada e camada oculta
hidden_weight = np.random.uniform(size=(inputLayerNeurons, hiddenLayerNeurons))
#Peso camada oculta e saida
output_weights = np.random.uniform(size=(hiddenLayerNeurons, outputLayerNeurons))

#taxa de aprendizado + épocas
lr = 0.5
epochs = 10000

#backpropagation
for epoch in range(epochs):
  hidden_layer_activation = np.dot(inputs, hidden_weight)
  hidden_layer_output = sigmoid(hidden_layer_activation)

  output_layer_activation = np.dot(hidden_layer_output, output_weights)
  predicted_output = sigmoid(output_layer_activation)

  # Cálculo do erro
  error = expected_output - predicted_output

    # Gradiente na saída
  d_predicted_output = error * sigmoid_derivative(predicted_output)

    # Erro na camada oculta
  error_hidden_layer = d_predicted_output.dot(output_weights.T)
  d_hidden_layer = error_hidden_layer * sigmoid_derivative(hidden_layer_output)

    # Atualização dos pesos
  output_weights += hidden_layer_output.T.dot(d_predicted_output) * lr
  hidden_weight += inputs.T.dot(d_hidden_layer) * lr

# 5 Teste
print("Resultados após o treinamento:")
for i in range(len(inputs)):
    pred = sigmoid(x=np.dot(sigmoid(np.dot(inputs[i], hidden_weight)), output_weights))
    print(f"Entrada: {inputs[i]} -> Predição: {pred[0]:.4f} (Esperado: {expected_output[i][0]})")
```

```python
#3ª versão
import numpy as np

# A Sigmóide é o "filtro" que dá o resultado, e a Derivada dela é a
#"ferramenta de medida" que diz o quanto o erro é urgente ou não.

# Sigmóide: transforma qualquer número em um valor entre 0 e 1.
# Ela é a "curva" que permite que a rede aprenda padrões complexos.
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivada da Sigmóide diz a inclinação da curva.
# usando no Backpropagation para saber "quanto" ajustar o peso.
def sigmoid_derivative(x):
    return x * (1 - x)

# --- 2.(A Tabela Verdade do XOR) ---

# Entradas
entradas = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Saídas Esperadas
saidas_reais = np.array([[0], [1], [1], [0]])

# --- 3. INICIALIZAÇÃO ---

# 2 neurônios entrada, 3 "camada escondida" e 1 de saída.
input_size = 2
hidden_size = 3
output_size = 1

# Inicializa com valores aleatórios pequenos.
# Pesos é a "importância" que a rede dá a cada conexão.
pesos_entrada_oculta = np.random.uniform(size=(input_size, hidden_size))
pesos_oculta_saida = np.random.uniform(size=(hidden_size, output_size))

# Taxa de aprendizado: o tamanho do passo que damos em direção ao acerto.
taxa_aprendizado = 0.3
epocas = 10000  # quantidade de vezes que vai praticar

# --- 4.APRENDIZADO ---

for i in range(epocas):

    # --- FASE 1: FORWARD PROPAGATION  ---
    # 1. Multiplica a entrada pelos pesos e aplica a sigmóide
    camada_oculta_ativacao = np.dot(entradas, pesos_entrada_oculta)
    camada_oculta_saida = sigmoid(camada_oculta_ativacao)

    # 2. Pega o resultado da camada oculta e leva para a saída final
    camada_saida_ativacao = np.dot(camada_oculta_saida, pesos_oculta_saida)
    predicao = sigmoid(camada_saida_ativacao)

    # --- FASE 2: BACKPROPAGATION  ---
    # 1. Calculamos o Erro:
    erro_saida = saidas_reais - predicao

    # 2. Calcula o Delta da Saída: (Erro * Inclinação da Curva)
    # Isso diz a direção da mudança necessária.
    delta_saida = erro_saida * sigmoid_derivative(predicao)

    # 3. Calcula o Erro na Camada Oculta:
    # O erro da saída é "repassado" para os neurônios de trás através dos pesos.
    erro_camada_oculta = delta_saida.dot(pesos_oculta_saida.T)
    delta_oculta = erro_camada_oculta * sigmoid_derivative(camada_oculta_saida)

    # --- FASE 3: ATUALIZA OS PESOS (O Aprendizado Real) ---
    # Ajusta os pesos multiplicando o que entrou pelo delta calculado.
    pesos_oculta_saida += camada_oculta_saida.T.dot(delta_saida) * taxa_aprendizado
    pesos_entrada_oculta += entradas.T.dot(delta_oculta) * taxa_aprendizado

# --- 5. TESTE FINAL ---
print("--- Resultado Final da Rede ---")
print(predicao)
```
